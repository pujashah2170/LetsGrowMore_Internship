#### Step-1 Read the image in RBG format

```python
from google.colab.patches import cv2_imshow
```

```python
image=cv2.imread('leopard.jpg')
plt.imshow(image)
```

```python
image.shape
```

#### Step-2 Converting the coloured image into grayscale image 

```python
gray=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
gray.shape
```

```python
plt.imshow(gray,cmap='gray')
```

#### Step-3 Inverting the grayscale image 

```python
inverted_image =255- gray
plt.imshow( inverted_image)
```

#### Step-4 Blurr the image 

```python
blurred = cv2.GaussianBlur(inverted_image, (21, 21), 0)
plt.imshow(blurred)
```

#### Step-5 Inverting the blurr image and by mixing the grayscale image by the inverted blurry image we can easily create the pencil sketch

```python
inverted_blurred = 255 - blurred
pencil_sketch = cv2.divide(gray, inverted_blurred, scale=248.0)
plt.imshow(pencil_sketch)
```
